```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
df=pd.read_csv('../data/basket_analysis.csv')
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
df_encoded = df.applymap(lambda x: 1 if x else 0)
```

```python
df_encoded.head()
```

```python
# Apply the Apriori algorithm to find frequent itemsets with a lower minimum support (0.2)
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

# Check if frequent_itemsets is empty
if frequent_itemsets.empty:
    print("No frequent itemsets found. Try lowering the min_support threshold.")
else:
    # Generate association rules from the frequent itemsets with a minimum lift of 1
    rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1, num_itemsets=2)

    # Display the resulting rules
    print(rules)
```

```python
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules.head(25)
```

```python
# Filter rules with lift greater than 1.2
rules_filtered = rules[rules['lift'] > 1.2]

# Display the filtered rules
print(rules_filtered)
```

```python
# Filter rules with support greater than 1.2
rules_filtered = rules[rules['support'] > 0.2]

# Display the filtered rules
print(rules_filtered)
```
