```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

### EDA

```python
df=pd.read_csv('data/hiring.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lr=LabelEncoder()
df['experience']=lr.fit_transform(df['experience'])
df
```

```python
df.isna().sum()
```

```python
df['test_score(out of 10)']=df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].mean())
```

```python
x=df.iloc[:,:-1].values
x
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)

from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)

y_pred=lr.predict(x_test)
y_pred
```

```python
print('coefficient:',lr.coef_)
print('intercept:',lr.intercept_)

from sklearn.metrics import r2_score
print('r2:',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Lasso
ls=Lasso(alpha=100)
ls.fit(x_train,y_train)
y_pred=ls.predict(x_test)
print('lasso coefficient:',ls.coef_)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print('mea:',mean_absolute_error(y_test,y_pred))
print('mse:',mean_squared_error(y_test,y_pred))
print('r2:',r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=88)
ridge.fit(x_train,y_train)
y_pred=ridge.predict(x_test)
print('ridge coefficient:',ridge.coef_)
```

```python
print('r2:',r2_score(y_test,y_pred))
```

# DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
```

```python
test_pred=dt.predict(x_test)
train_pred=dt.predict(x_train)
```

```python
print('testr2:',r2_score(y_test,test_pred))
print('trainr2:',r2_score(y_train,train_pred))
```
