# Imbalanced Datasets

When the target variable (e.g., TenYearCHD) has an unequal distribution, where one class (e.g., class 0) has significantly more instances than the other class (e.g., class 1), the dataset is considered imbalanced. This imbalance can lead to the model being biased toward predicting the majority class, which reduces its ability to correctly predict the minority class.

Ways to Handle Imbalanced Datasets:

# 1.Over-sampling (Increase Minority Class):

In over-sampling, we artificially increase the number of instances in the minority class (e.g., 1) by duplicating instances or generating synthetic samples.
SMOTE (Synthetic Minority Over-sampling Technique) is one common approach where synthetic data points are generated based on the nearest neighbors of minority class samples.
Advantages of Over-sampling:

The model gets more examples of the minority class, which helps it learn the patterns for that class.
It can lead to better classification performance for the minority class.
Disadvantages of Over-sampling:

It may lead to overfitting since the model may learn the repeated instances or synthetic samples too well.
The training time may increase due to the added samples.

### Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, confusion_matrix, accuracy_score, classification_report
```

```python
data=pd.read_csv('data/framingham.csv')
data.head()
```

### Data preprocessing

```python
# Handle missing values column by column
data['education'] = data['education'].fillna(data['education'].mean())
data['cigsPerDay'] = data['cigsPerDay'].fillna(data['cigsPerDay'].mean())
data['BPMeds'] = data['BPMeds'].fillna(data['BPMeds'].mean())
data['totChol'] = data['totChol'].fillna(data['totChol'].mean())
data['BMI'] = data['BMI'].fillna(data['BMI'].mean())
data['heartRate'] = data['heartRate'].fillna(data['heartRate'].mean())
data['glucose'] = data['glucose'].fillna(data['glucose'].mean())

# Verify missing values are resolved
print(data.isnull().sum())
```

### Correlation Analysis

```python
correlation_matrix = data.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

### Data Visualization

```python
# Scatter plot for Age vs. Total Cholesterol (totChol)
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x="age", y="totChol", hue="TenYearCHD", palette="Set1")
plt.title("Age vs Total Cholesterol (totChol) by TenYearCHD")
plt.xlabel("Age")
plt.ylabel("Total Cholesterol (totChol)")
plt.legend(title="TenYearCHD", loc="upper left")
plt.show()

# Boxplot to visualize distributions and detect outliers
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[["age", "totChol", "BMI", "sysBP", "diaBP", "heartRate", "glucose"]])
plt.title("Boxplot of Numerical Features")
plt.xticks(rotation=45)
plt.show()

# Lmplot to visualize relationships
sns.lmplot(data=data, x="age", y="sysBP", hue="TenYearCHD", height=6, aspect=1.5)
plt.title("Age vs SysBP by TenYearCHD")
plt.show()

sns.lmplot(data=data, x="BMI", y="glucose", hue="TenYearCHD", height=6, aspect=1.5)
plt.title("BMI vs Glucose by TenYearCHD")
plt.show()
```

### Label Encoding

```python
le = LabelEncoder()
data['education'] = le.fit_transform(data['education'].astype(str))
```

### Splitting Data into Inputs (X) and Target (Y)

```python
X = data.drop(columns=["TenYearCHD"])
y = data["TenYearCHD"]
```

### Scaling Inputs

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### Over Sampling

```python
data["TenYearCHD"].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
```

```python
over=SMOTE(random_state=2)
X_os,y_os=over.fit_resample(X_scaled,y)
```

```python
y_os.value_counts()
```

### Splitting Data into Train and Test Sets

```python
X_train, X_test, y_train, y_test = train_test_split(X_os, y_os, test_size=0.2, random_state=42)
```

### Logistic Regression Modeling

```python
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
```

### Predictions

```python
y_pred = logistic_model.predict(X_test)
y_pred
```

### Visualization of Actual & Predicted Values

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, label="Actual Values", alpha=0.6)
plt.scatter(range(len(y_pred)), y_pred, label="Predicted Values", alpha=0.6)
plt.legend()
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(len(y_test)), y_test, label="Actual Values", alpha=0.6, color='blue', linestyle='-', marker='o')
plt.plot(range(len(y_pred)), y_pred, label="Predicted Values", alpha=0.6, color='red', linestyle='-', marker='x')
plt.legend()
plt.title("Actual vs Predicted Values")
plt.show()
```

```python
results = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
sns.pairplot(results)
plt.show()
```

### Creating DataFrame to View Results

```python
results_df = pd.DataFrame({"Actual": y_test, "Predicted": y_pred})
print(results_df.head())
```

### Model evaluation

```python
# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

```python
# Accuracy Score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy*100,"%")
```

```python
# Classification Report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)
```
